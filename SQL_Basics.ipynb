9.

   - A Primary key uniquely identifies each record in a table and cannot contain NULL values.

   - A Foreign Key is used to link one table to another and refers to the primary key of another table.

10.

   - Constraints are rules applied to table columns to maintain data accuracy and integrity.

   - Examples include PRIMARY KEY, NOT NULL, UNIQUE, CHECK and FOREIGN KEY.
